# Check which sample you're working on!!!

```python
import pandas as pd
```

```python
chrRange = range(1,23)
dfList = []
for CHR in chrRange:
    CHR = str(CHR)
    PSI_Path = "/storage/ydong/data/chrom/tibialNerve/chr" + CHR + "_PSI.csv"
    PSI_df = pd.read_csv(PSI_Path, index_col = "Unnamed: 0")
    dfList.append(PSI_df)
```

```python
PSI_total = pd.concat(dfList,axis = 1)
```

```python
PSI_total
```

```python
PSI_total.to_csv("/storage/ydong/data/chrom/tibialNerve/all_PSI.csv")
```

# Trim out the PSI columns with median <= 0 or >= 100

```python
PSI_total = pd.read_csv("/storage/ydong/data/chrom/lung/all_PSI.csv")
PSI_total = PSI_total.set_index("Unnamed: 0")
```

```python
#only select those columns with median < 100 and > 0 
meds = PSI_total.median(axis = 0) 
PSI_selected = PSI_total[meds[(meds < 100) & (meds > 0)].keys()]
```

```python
PSI_selected
```

```python
PSI_selected.to_csv("/storage/ydong/data/chrom/lung/all_PSI_trimmed.csv")
```

```python
df = pd.read_csv("/storage/ydong/data/gtexRNA/gencode.v26.GRCh38.genes.gtf",sep = "\t", skiprows = 5)
```

```python
df
```

```python
#extract exon annotation only
idx = pd.IndexSlice
df = df.loc[idx[:,:,"exon"],:]
```

```python
df
```

```python
# chromosome number
chromNum = list(df.index.get_level_values(0))
# transform from "chrX" to "X"
for i in range(len(chromNum)):
    chromNum[i] = chromNum[i].split("chr")[-1]
    
# start position
start = list(df.index.get_level_values(3))
# end position
end = list(df.index.get_level_values(4))
# forward/reverse strand
strand = list(df.index.get_level_values(6))

#extract exonID from the values
exonID = list(df.values)
for i in range(len(exonID)):
    curr = exonID[i][0]
    curr = curr.strip().split(";")[-3].split("\"")[1]
    exonID[i] = curr
```

```python
df = pd.DataFrame (list(zip(exonID,chromNum,start,end,strand)),columns= ["probe.id","probe.chr","probe.start","probe.stop","probe.strand"])
df["gene.id"] = df["probe.id"].apply(lambda x:x.split("_")[0])
df
```

```python
df.to_csv("/storage/ydong/data/gtexRNA/gencode_gene_annotations_GRCh38.csv")
```

```python
import pandas as pd
```

```python
chrRange = range(1,23)
dfList = []
for CHR in chrRange:
    CHR = str(CHR)
    PSI_Path = "/storage/ydong/data/chrom/chr" + CHR + "_PSI.csv"
    PSI_df = pd.read_csv(PSI_Path, index_col = "Unnamed: 0")
    dfList.append(PSI_df)
```

```python
PSI_total = pd.concat(dfList,axis = 1)
```

```python
PSI_total
```

```python
PSI_total.to_csv("/storage/ydong/data/chrom/all_PSI.csv")
```
